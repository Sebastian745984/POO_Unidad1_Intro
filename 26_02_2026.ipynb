```
class Libro:
  def __init__(self, isbn, titulo, autor, categoria, precio):
    self.isbn = isbn
    self.titulo = titulo
    self.autor = autor
    self.categoria = categoria
    self.precio = precio

  def mostrar_info(self):
      estado = "Prestado" if self.prestado else "Disponible"
      print(f"[{self.isbn}] {self.titulo} - {self.autor} | {self.categoria} | ${self.precio} | {estado}")

class InventarioLibro:
  def __init__(self, libro, cantidad):
    self.libro = libro
    self.total = cantidad
    self.prestados = 0
    self.vendidos = 0

  def disponibles(self):
    return self.total - self.prestados - self.vendidos

  def prestar(self):
      if self.disponibles() > 0:
           self.prestados += 1
           return True
      return False
  def devolver(self):
      if self.prestados > 0:
           self.prestados -= 1
           return True
      return False

  def vender(self):
      if self.disponibles() > 0:
           self.vendidos += 1
           return True
      return False

  def mostrar_estado(self):
      l = self.libro
      print(f"[{l.isbn}] {l.titulo} - {l.autor} | Total: {self.total} | \
      Disponibles: {self.disponibles()} |  Prestados: {self.prestados} | Vendidos: {self.vendidos}")

class Estudiante:
  def __init__(self, num_control, nombre, carrera, fecha_nacimiento, telefono, correo):
      self.num_control = num_control
      self.nombre = nombre
      self.carrera = carrera
      self.fecha_nacimiento = fecha_nacimiento
      self.telefono = telefono
      self.correo = correo
  def mostrar_info(self):
      print(f"[{self.num_control}] {self.nombre} - {self.carrera} | Nac: {self.fecha_nacimiento} | Tel: {self.telefono} | Correo: {self.correo}")

class Prestamo:
      contador_prestamos = 1
      def __init__(self, num_control, isbn, fecha_inicio, fecha_termino):
        self.folio = Prestamo.contador_prestamos
        Prestamo.contador_prestamos += 1
        self.num_control = num_control
        self.isbn = isbn
        self.fecha_inicio = fecha_inicio
        self.fecha_termino = fecha_termino

        def mostrar_info(self):
          print(f"Folio: {self.folio} | Estudiante: {self.num_control} | Libro ISBN: {self.isbn} | Del {self.fecha_inicio} al {self.fecha_termino}")

class Venta:
  contador_ventas = 1
  def __init__(self, isbn, num_control, fecha, cantidad, importe):
    self.folio_ventas = Venta.contador_ventas
    Venta.contador_ventas += 1
    self.isbn = isbn
    self.num_control = num_control
    self.fecha = fecha
    self.cantidad = cantidad
    self.importe = importe

  def mostrar_info(self):
    print(f"Folio: {self.folio_ventas} | ISBN: {self.isbn} | Comprador: {self.num_control} | Fecha: {self.fecha} | Cantidad: {self.cantidad} | Importe: {self.importe}")

inventarios = []
estudiantes = []
prestamos = []
ventas = []

# The 'libros' list was redundant as inventory items store the books. Removed it.
# libros = []

def alta_libro():
    print("\n--- ALTA DE LIBRO ---")
    isbn = input("ISBN: ")
    for inv in inventarios:
      if inv.libro.isbn == isbn:
        print("El libro ya existe con ese ISBN")
        return
    titulo = input("Titulo: ")
    autor = input("Autor: ")
    categoria = input("Categoria: ")
    try:
        precio = float(input("Precio: "))
    except ValueError:
      print("Precio Invalido")
      return
    try:
        cantidad = int(input("Cantidad de libros: ")) # Added input for cantidad
    except ValueError:
        print("Cantidad invalida")
        return
    nuevo_libro = Libro(isbn, titulo, autor, categoria, precio)
    inventario_item = InventarioLibro(nuevo_libro, cantidad)
    inventarios.append(inventario_item)
    print("Libro dado de alta correctamente")

    def mostrar_inventario():
      print("\n--- INVENTARIO DE LIBROS ---")
      if not inventarios:
          print("No hay libros en el inventario")
      else:
            for item in inventarios:
                item.mostrar_estado()


def alta_estudiante():
  print("\n -- Alta de estudiante ---")
  num_control = input("Numero de control: ")
  for est in estudiantes:
    if est.num_control == num_control:
      print("El estudiante ya existe con ese numero de control")
      return
  nombre = input("Nombre: ")
  carrera = input("Carrera: ")
  fecha_nacimiento = input("Fecha de nacimiento (dd/mm/aaaa): ")
  telefono = input("Telefono: ")
  correo = input("Correo: ")
  nuevo_estudiante = Estudiante(num_control, nombre, carrera, fecha_nacimiento, telefono, correo)
  estudiantes.append(nuevo_estudiante)
  print("Estudiante dado de alta correctamente")

def consultar_libro():
    print("\n--- CONSULTA DE LIBRO POR ISBN ---")
    isbn = input("Ingrese el ISBN del libro: ")
    for inv_item in inventarios: # Iterate inventarios instead of libros
        if inv_item.libro.isbn == isbn: # Access book info through inventario item
             inv_item.libro.mostrar_info()
             return
    print("Libro no encontrado")

def consultar_estudiante():
    print("\n--- CONSULTA DE ESTUDIANTE POR NUMERO DE CONTROL ---")
    num_control = input("Ingrese el numero de control del estudiante: ")
    for est in estudiantes:
        if est.num_control == num_control:
             est.mostrar_info()
             return
    print("Estudiante no encontrado")

def mostrar_todos_libros():
    print("\n--- LISTADO DE TODOS LOS LIBROS ---")
    if not inventarios: # Check inventarios instead of libros
        print("No hay libros registrados")
    else:
        for inv_item in inventarios: # Iterate inventarios
            inv_item.mostrar_estado() # Display full inventory status

def mostrar_todos_estudiantes():
    print("\n--- LISTADO DE TODOS LOS ESTUDIANTES ---")
    if not estudiantes:
        print("No hay estudiantes registrados")
    else:
        for est in estudiantes:
            est.mostrar_info()

def realizar_prestamo():
    print("\n--- REALIZAR PRESTAMO ---")
    num_control = input("Numero de control del estudiante: ")
    isnt = input("ISBN del libro a prestar: ") # 'isnt' variable name was inconsistent

    estudiante = next((e for e in estudiantes if e.num_control == num_control), None)
    if not estudiante:
        print("Estudiante no encontrado")
        return

    inv_item = next((inv for inv in inventarios if inv.libro.isbn == isnt), None) # Find InventarioLibro
    if not inv_item:
        print("Libro no encontrado en el inventario")
        return

    # Check if the book can be lent
    if inv_item.prestar():
        fecha_inicio = input("Fecha de inicio del prestamo (dd/mm/aaaa): ")
        fecha_termino = input("Fecha de termino del prestamo (dd/mm/aaaa): ")
        prestamo = Prestamo(num_control, isnt, fecha_inicio, fecha_termino) # Pass 'isnt'
        prestamos.append(prestamo)
        print(f"Prestamo registrado con folio {prestamo.folio}.")
    else:
        print("Libro no disponible para prestamo.")

# Removed nested mostrar_prestamos() function as a global one already exists.

def devolver_libro():
    print("\n--- DEVOLVER LIBRO ---")
    try:
        folio = int(input("Folio del prestamo a devolver: "))
    except ValueError:
        print("Folio invalido")
        return

    # Corrected indentation for prestamo variable
    prestamo = next((p for p in prestamos if p.folio == folio), None)
    if not prestamo:
        print("Prestamo no encontrado")
        return

    inv_item = next((inv for inv in inventarios if inv.libro.isbn == prestamo.isbn), None)
    if inv_item:
        if inv_item.devolver(): # Use InventarioLibro's devolver method
            prestamos.remove(prestamo)
            print("Libro devuelto correctamente")
        else:
            print("Error al devolver el libro (no estaba prestado o ya fue devuelto).")
    else:
        print("Libro no encontrado en el inventario.")

def registrar_venta():
    print("\n--- REGISTRAR VENTA ---")
    isbn = input("ISBN del libro vendido: ")
    inv_item = next((inv for inv in inventarios if inv.libro.isbn == isbn), None) # Find InventarioLibro
    if not inv_item:
        print("Libro no encontrado en el inventario")
        return

    try:
        cantidad_a_vender = int(input(f"Cantidad a vender (disponibles: {inv_item.disponibles()}): "))
        if cantidad_a_vender <= 0:
            print("La cantidad debe ser un numero positivo.")
            return
        if cantidad_a_vender > inv_item.disponibles():
            print("No hay suficientes libros disponibles para la venta.")
            return
    except ValueError:
        print("Cantidad invalida")
        return

    num_control = input("Numero de control del comprador (999999 si es publico general): ")
    if num_control != "999999":
      estudiante = next((e for e in estudiantes if e.num_control == num_control), None)
      if not estudiante:
        print("Estudiante no encontrado")
        return

    fecha = input("Fecha de la venta (dd/mm/aaaa): ")
    importe = inv_item.libro.precio * cantidad_a_vender

    # Perform the sale by calling vender multiple times
    for _ in range(cantidad_a_vender):
        inv_item.vender()

    venta = Venta(isbn, num_control, fecha, cantidad_a_vender, importe)
    ventas.append(venta)
    # Removed libros.remove(libro) as inventory is managed by InventarioLibro
    print(f"Venta registrada con folio {venta.folio_ventas}.")

# Removed nested mostrar_historial_ventas() function as a global one already exists.


def mostrar_menu():
  print("\n--- SISTEMA DE LIBRERIA ESCOLAR ---")
  print("1. Dar de alta un Libro")
  print("2. Dar de alta estudiante")
  print("3. Colsultar libro por ISBN")
  print("4. Consultar estudiante por numero de control")
  print("5. Mostrar todos los libros")
  print("6. Mostrar todos los estudiantes")
  print("7. Realizar prestamo")
  print("8. Mostrar prestamos")
  print("9. Devolver libro")
  print("10. Registrar ventas")
  print("11. Mostrar historial de Ventas")
  print("12. Salir del sistema")

def iniciar_sistema():
      while True:
        mostrar_menu()
        opcion = input("Seleccione una opcion: ")
        if opcion == '1':
          alta_libro()
        elif opcion == '2':
          alta_estudiante()
        elif opcion == '3':
            consultar_libro()
        elif opcion == '4':
            consultar_estudiante()
        elif opcion == '5':
            mostrar_todos_libros()
        elif opcion == '6':
            mostrar_todos_estudiantes()
        elif opcion == '7':
            realizar_prestamo()
        elif opcion == '8':
            mostrar_prestamos_global()
        elif opcion == '9':
            devolver_libro()
        elif opcion == '10':
            registrar_venta()
        elif opcion == '11':
            mostrar_historial_ventas_global()
        elif opcion == '12':
            print("Saliendo del sistema...")
            break
        else:

          print("Opcion invalida. Intente de nuevo")

def mostrar_prestamos_global():
    print("\n--- LISTADO DE PRESTAMOS ---")
    if not prestamos:
        print("No hay prestamos registrados")
    else:
        for p in prestamos:
            p.mostrar_info() # Call the object's method

def mostrar_historial_ventas_global():
    print("\n--- HISTORIAL DE VENTAS ---")
    if not ventas:
        print("No hay ventas registradas")
    else:
        for v in ventas:
            v.mostrar_info() # Call the object's method

iniciar_sistema()
```
